```python
financial=Data.loc[:,['Sales','Quantity','Discount','Profit']]
g = sns.PairGrid(financial)
g.map_upper(sns.scatterplot,color='purple')
g.map_lower(sns.scatterplot, color='brown')
g.map_diag(plt.hist)
```
